```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hh_data = "..Data/processed/henry_hub_spot.csv"

wti_data = "../Data/processed/wti_spot.csv"

hh = pd.read_csv(hh_data, parse_dates=["date"])
wti = pd.read_csv(wti_data, parse_dates=["date"])

hh = hh.dropna(subset=["HH_spot"])
df = pd.merge(hh, wti, on="date", how="inner")
df = df.rename(columns={"HH_spot": "HH", "WTI_spot": "WTI"})
df = df.set_index("date")
df = df.sort_index()

print(df)
```

```python
df.plot(y=["HH", "WTI"], figsize=(12, 5), title="Henry Hub vs WTI Spot Prices")
plt.grid(True)
plt.show()
```

```python
weather_agg_csv = "/Users/rezapirbhai/HenryHubPredicting/Data/processed/weather_agg.csv"
weather_agg = pd.read_csv(weather_agg_csv, parse_dates=["date"])
df = df.drop(columns=["CDD", "HDD"], errors="ignore")
df = pd.merge(df, weather_agg, on="date", how="inner")
print(df)

output_csv = "/Users/rezapirbhai/HenryHubPredicting/Data/merged_data.csv"
df.to_csv(output_csv, index=False)
```
